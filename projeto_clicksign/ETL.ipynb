```python
#conforme visto no dataframe.dtypes, os dados relativo às colunas age, fnlwgt, education-num  e hours-per-week não
#estão seguindo os dados com os quais foram estabelecidos da forma acima, isto é, todos precisam estar na forma de valores inteiros
# para investigar melhor o porquê disso, criou-se uma função buscando armazenar os valores que não são correspondentes

def validate_ints(df_indexes):
    """o occurences types irá armazenar os dados inválidos, a estrutura segue a seguinte logica:
    o número da chave representa o número da linha na qual está este dados, e o valor é uma lista contendo tal dado e sua respectiva coluna"""
    global be_integers
    be_integers=df_indexes
    occurences_types={}
    for col in be_integers:
        for lin in range(total_row):
            val=data_numpy[lin][col]
            char_float= isinstance(data_numpy[lin][col],float)
            if char_float:
                if np.isnan(val)==True or int(val)!=val:
                    occurences_types[lin]=[data_numpy[lin][col], header[col]]        
            else:
                char_aval=str(val).replace(' ','')
                if char_aval.isnumeric() != True:
                    occurences_types[lin]=[data_numpy[lin][col], header[col] ] 
    return occurences_types

        
object1=validate_ints([0,2,4,12])
print(object1)
```

```python
#Já no capital-gain, de forma a ser analisada unicamente, ele , ao contrário acima que não podia ser float, precisa ser 
lista_indexs_gain=[]
#utilizei as expressões regulares para identificar se a string possui o formato de float.
for value in dataframe['capital-gain']:
    x=str(value).replace(' ','')
    pattern=r"[0]?[1-9]*(.)?[0-9]*"
    y=re.search(pattern, x)
    v=y.span()
    if v==(0,1) and x.isnumeric() == False:
        #quando a tupla vai de zero a 1, é preciso fazer uma validação pra ver se é só uma letra, se sim, precisamos adicionar
        lista_indexs_gain.append(x)
            
object2=lista_indexs_gain
print(object2)
#existe um elemento A na coluna referente à capital-gain
```

```python
#Fora isso, realizou-se outra avaliação pra saber se ocorrem dados que distoam dos valores estabelidos pelo estudo de caso
#Para isso,os dados do arquivo de descrição foram copiados gerando uma lista, segundo está abaixo, retirando os campos que já foram analisados:
#OBS: os dados foram copiados para não correr o risco de algum caractere ser apagado

remaining_evaluation=["workclass     : Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay, Never-worked",
"education     : Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool",
"marital-status: Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse",
"occupation    : Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces",
"relationship  : Wife, Own-child, Husband, Not-in-family, Other-relative, Unmarried",
"race          : White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black",
"sex           : Female, Male",
"native-country: United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands",
"class         : >50K, <=50K" ]

c=0
for linh in remaining_evaluation:
    print(f'posição ponto e virgula da linha {c} -> {linh.index(":")}')
    c+=1
    #identificar onde fica os dois pontos e começar a separar as strings a partir dele
    #já que todos ficaram alinhando, dando 14, a posição ini sera 14+1
```

```python
pos_ini=15
remanescent=[]
for lin in remaining_evaluation:
    remanescent.append((lin[pos_ini+1:].replace(' ','').lower()).split(','))
```

```python
#após a separação dos dados, a lógica é parecida com a função validate_type, porém será avaliado se este dado está contido em cada linha acima.Se ele não estiver, irei armazenar o mesmo. 

def validate_strings(*args):
    """valid_columns é o objeto que irá representar as colunas finais que precisam ser validadas, 
    que será a diferença entre o total de colunas, menos as que já foram validadas acima e as que já tinham valores contínuos já estabelecidos"""
   
    tot_indexes_dataframe=set(list(i for i in range((total_col))))
    col_be_integers=set(be_integers)
    col_be_gain=set([10,11])  #refere-se a coluna do capital-gain e capital-loss, que não precisa ser avaliada 
    valid_columns=list(tot_indexes_dataframe.difference(col_be_integers.union(col_be_gain)))
    align_indexes={}
    valid_list=[]
    for v in range(len(valid_columns)):
        align_indexes[valid_columns[v]]= v
    for col in valid_columns:
        for lin in range(total_row):
            data_mod=str(data_numpy[lin][col].replace(' ','')).lower()
            #validação avaliando se o o elemento está dentro dos valores estabelecidos no remaining_evaluation
            if (data_mod) not in remanescent[align_indexes[col]]:
                valid_list.append(data_numpy[lin][col])
            else:
            #outra validação, pois pode ser que o elemento esteja dentro dos valores estabelecidos, porém não tem o mesmo comprimento
                index=remanescent[align_indexes[col]].index(data_numpy[lin][col].replace(' ','').lower())
                if len(data_mod) != len(remanescent[align_indexes[col]][index]):
                    valid_list.append(data_numpy[lin][col])
    return valid_list
```

```python
#avaliando os elementos unicos da função validate_type2
elements=validate_strings()
uniques_elem=[]
for elem in elements:
    if elem not in uniques_elem:
        uniques_elem.append(elem)
object3=uniques_elem        
print(object3)
```

```python
#Busquei criar uma função pra realizar o processo de visualização de dados duplicados 
#possibilidade de melhoria talvez com recursão e memoização

def analyze_duplicate():
    lit,duplicated_indexes=[],[]
    dici={}
    #a identificação das linhas duplicadas se baseia na leitura de cada linha, inserindo-a em uma lista
    #Dessa forma, utilizando o comando index, ele nos mostra o primeiro elemento dessa lista
    #Ou seja, utilizando essa lógica, se o indice dessa lista está aparecendo antes do indice tratado no loop
    #o arquivo está duplicado
    list_numpy=[list(i) for i in data_numpy]
   
    for k,v  in enumerate(list_numpy):
        if list_numpy.index(v) != k:
            duplicated_indexes.append([list_numpy.index(v),k])
            
    
    return duplicated_indexes


obj_dupli_lin=analyze_duplicate()
print((obj_dupli_lin))
```

#####  método Z-Score

```python
#o método Z-Score relaciona se baseia na investigação de um certo dado em como ele se relaciona com a média e o desvio padrão do grupo com o qual este dado está inserido
# A fórmula é representada da seguinte forma: z = (x – μ)/σ
#onde o x é a amostra avaliada, mi a média e o sigma o desvio padrão
#Para identificação do outlier, este valor não pode passar de 3 ou -3

#este método será aplicado às colunas numéricas que fazem sentido, como é o caso de age, fnlwgt, education-num,capital-gain,capital-loss, e horas trabalhadas por semana

#iremos construir uma função para avaliar os outliers, inicialmente 

```

```python
def analyze_outliers(datafram,colunm_index):
    #para tratar os dados antes de passar pelo loop, quando passar por um NaN
    try:
        data_filter=datafram[header[colunm_index]].replace({np.nan:0})
        transform_int=list(map(lambda x:int(x),data_filter))
        media=sum(transform_int)/len(data_filter)
        desv_pad=np.std(transform_int)  
        validador=3
        outliers={}
        
        "os valores serão armazenados em um dicionário representando o indice:[valor do z, valor sob este indice]"
        
        for lin in range(total_row):
            value=transform_int[lin]
            z=(value - media)/desv_pad
            if z > validador or z<-3:
                outliers[lin]=[z,value]
        return outliers 
    except ValueError:
         print("O dado não pode ser convertido pra inteiro ")
                
analyze_outliers(dataframe_test,4)
```

```python
#no caso da idade,utilizando o algoritmo do outlier acima, os valores que ultrapassaram +3 podem ser ainda considerados como validos
#Já o outlier -3 não apareceu ,não resultando em nenhum dado abaixo de zero(idade logicamente não existente)
#conforme mostra abaixo:
analyze_outliers(dataframe_test, 0)
```
